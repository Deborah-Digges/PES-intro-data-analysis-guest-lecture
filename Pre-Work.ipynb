# Types of Data

## Categorical/Qualitative: 

- data doesn't have any numerical values, only "categories"
- A good visual is a bar chart for comparison
- Pie chart to see the contribution of each class to the whole

### Which is the most popular sector in which people work?

```python
df.workclass.value_counts().plot.bar();
```

```python
df.workclass.value_counts(normalize=True).plot.pie();
```

### What does the education of people look like?

```python
df.education.value_counts(normalize=True).plot.barh();
```

### What is the marital status of people?

```python
df.marital_status.value_counts().plot.bar();
```

### Occupation?

```python
df.occupation.value_counts().plot.bar();
```

```python
df.race.value_counts().plot.bar();
```

## Quantitative data ("Numbers")

### What is the age distribution of people?

```python
df.age.plot.hist();
```

### What is the average age of people?

```python
df.age.mean()
```

### What is the median(2nd quartile) age of people?

```python
df.age.median()
```

### What is the 1st quartile?

```python
df.age.quantile(0.25)
```

### What is the third quartile?

```python
df.age.quantile(0.75)
```

### Is there a way to show the median, 1st and 3rd quartiles together?

```python
df.age.plot.box();
```

### What is the distribution of the number of years of education?

```python
df.education_num.plot.hist();
```
